```python
import pandas as pd
```

```python
df = pd.read_csv('ecommerce_sales_data.csv')
```

```python
df.columns = df.columns.str.replace(' ' , '-')
```

```python
df['Order-Date'] = df['Order-Date'].astype('date64[pyarrow]')
```

```python
df['Year'] = df['Order-Date'].dt.year
```

```python
df['Month'] = df['Order-Date'].dt.month
```

```python
df['Day'] = df['Order-Date'].dt.day
```

```python
df.sample()
```

```python
df.drop(columns='Order-Date' , inplace=True)
```

```python
df.sample()
```

```python
df['Product-Name'].unique()
```

```python
X = df.drop(columns='Profit')

y = df['Profit']
```

```python
num_cols = df.select_dtypes(include='number').columns
cat_cols = df.select_dtypes(exclude='number').columns
```

```python
from sklearn.model_selection import train_test_split

X_train , X_test , y_train , y_test = train_test_split(X , y , random_state=42 , train_size= 0.8  ,shuffle=True)
```

```python
from sklearn.preprocessing import RobustScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.tree import DecisionTreeRegressor
from category_encoders.binary import BinaryEncoder
```

```python
num_cols
```

```python
cat_cols
```

```python
preprocessor = ColumnTransformer([
    ('scalling' , RobustScaler(),['Sales', 'Year' , 'Month' ,'Day']),
    ('encoding' , BinaryEncoder() , ['Product-Name' , 'Category' ,'Region'])
])
```

```python
pipeline = Pipeline([
    ('Preprocessor' , preprocessor),
    ('model' , DecisionTreeRegressor(min_samples_split=75))
])
```

```python
pipeline.fit(X_train , y_train)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

y_predict_train = pipeline.predict(X_train)
y_predict_test = pipeline.predict(X_test)

# MSE
mse_train = mean_squared_error(y_train, y_predict_train)
mse_test = mean_squared_error(y_test, y_predict_test)

# RMSE
# rmse_train = mean_squared_error(y_train, y_predict_train, squared=False)
# rmse_test = mean_squared_error(y_test, y_predict_test, squared=False)

# MAE
mae_train = mean_absolute_error(y_train, y_predict_train)
mae_test = mean_absolute_error(y_test, y_predict_test)

# R2 SCORE
r2_train = r2_score(y_train, y_predict_train)
r2_test = r2_score(y_test, y_predict_test)

print("MSE Train:", mse_train)
print("MSE Test:", mse_test)
print("*"*50)
# print("RMSE Train:", rmse_train)
# print("RMSE Test:", rmse_test)
print("MAE Train:", mae_train)
print("MAE Test:", mae_test)
print("*"*50)
print("R2 Score Train:", r2_train)
print("R2 Score Test:", r2_test)
```

```python
import pandas as pd

# 1 — Get feature names from preprocessor
feature_names = pipeline.named_steps['Preprocessor'].get_feature_names_out()

# 2 — Get model feature importances
importances = pipeline.named_steps['model'].feature_importances_

# 3 — Build DataFrame
fi_df = pd.DataFrame({
    'Feature': feature_names,
    'Importance': importances
})

# 4 — Sort and show
fi_df = fi_df.sort_values(by='Importance', ascending=False)
fi_df.reset_index(drop=True, inplace=True)

print(fi_df)
```
